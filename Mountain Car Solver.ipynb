```python
import tensorflow as tf
import numpy as np

sess = tf.Session()

num_inputs = 2
num_classes = 3
lr = 1e-3
b = 0

x = tf.placeholder(tf.float32, [None, num_inputs])
z = tf.nn.l2_normalize(x, dim=1)
#W1 = tf.Variable(tf.zeros((num_inputs, num_classes)), tf.float32)
l1 = tf.contrib.layers.fully_connected(z, 30)
l2 = tf.nn.relu(l1)

l3 = tf.contrib.layers.fully_connected(l2, num_classes)#tf.matmul(x, W1)
probs = tf.nn.softmax(l3)

y = tf.placeholder(tf.int32, shape=(None))
one_hot_y = tf.one_hot(y, num_classes)

A = tf.placeholder(tf.float32, shape=(None))

q = tf.reduce_sum(tf.reduce_sum(tf.log(probs)*one_hot_y, axis=1)*A)

optimizer = tf.train.GradientDescentOptimizer(lr)
train_op = optimizer.minimize(-q)
    
```
